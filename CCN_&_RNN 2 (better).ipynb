```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, Dropout, BatchNormalization, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
import talib
import logging

print("GPU Available: ", tf.config.list_physical_devices('GPU'))
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')


def create_features(data):
    logging.info(f"Creating features. Input data shape: {data.shape}")
    features = pd.DataFrame(index=data.index)

    # Basic features
    features['returns'] = data['close'].pct_change()
    features['log_return'] = np.log(data['close'] / data['close'].shift(1))
    features['volatility'] = features['returns'].rolling(window=20, min_periods=1).std()
    features['volume_change'] = data['volume'].pct_change()

    # Technical indicators
    features['rsi'] = talib.RSI(data['close'])
    features['macd'], _, _ = talib.MACD(data['close'])
    features['adx'] = talib.ADX(data['high'], data['low'], data['close'])
    features['cci'] = talib.CCI(data['high'], data['low'], data['close'])
    features['obv'] = talib.OBV(data['close'], data['volume'])

    # Moving averages
    features['sma_10'] = data['close'].rolling(window=10, min_periods=1).mean()
    features['sma_30'] = data['close'].rolling(window=30, min_periods=1).mean()
    features['ema_10'] = data['close'].ewm(span=10, adjust=False, min_periods=1).mean()
    features['ema_30'] = data['close'].ewm(span=30, adjust=False, min_periods=1).mean()

    # Bollinger Bands
    features['bb_upper'], features['bb_middle'], features['bb_lower'] = talib.BBANDS(data['close'])

    # Replace infinite values with NaN
    features = features.replace([np.inf, -np.inf], np.nan)

    # Fill NaN values
    features = features.fillna(method='ffill').fillna(method='bfill').fillna(0)

    logging.info(f"Features created. Output shape: {features.shape}")
    logging.info(f"Columns: {features.columns.tolist()}")
    logging.info(f"Non-null counts:\n{features.count()}")

    return features

def prepare_data(features, sequence_length=30, scaler=None):
    logging.info(f"Preparing data. Input features shape: {features.shape}")

    # Filter trading hours and create signal
    mask = features['returns'].abs() > 1e-8  # Use a small threshold instead of exact zero
    logging.info(f"Number of trading hours: {mask.sum()} out of {len(mask)}")

    X = features.drop(['returns', 'log_return'], axis=1)[mask]
    y = (features['returns'][mask] > 0).astype(int)

    logging.info(f"After filtering trading hours - X shape: {X.shape}, y shape: {y.shape}")

    if X.empty or y.empty:
        logging.warning("No valid data after preprocessing")
        return None, None, None

    # Scale the features
    if scaler is None:
        scaler = RobustScaler()
        X_scaled = scaler.fit_transform(X)
    else:
        X_scaled = scaler.transform(X)

    logging.info(f"After scaling - X_scaled shape: {X_scaled.shape}")

    # Create sequences
    X_seq, y_seq = [], []
    for i in range(len(X_scaled) - sequence_length):
        X_seq.append(X_scaled[i:i+sequence_length])
        y_seq.append(y.iloc[i+sequence_length])

    X_seq = np.array(X_seq)
    y_seq = np.array(y_seq)

    logging.info(f"Final sequences - X_seq shape: {X_seq.shape}, y_seq shape: {y_seq.shape}")

    if X_seq.size == 0 or y_seq.size == 0:
        logging.warning("No valid sequences created")
        return None, None, None

    return X_seq, y_seq, scaler


def create_model(input_shape):
    model = Sequential([
        Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape),
        BatchNormalization(),
        Dropout(0.2),
        Conv1D(filters=128, kernel_size=3, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        LSTM(100, return_sequences=True),
        Dropout(0.2),
        LSTM(100),
        Dropout(0.2),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

def evaluate_model(y_true, y_pred):
    y_true_class = np.argmax(y_true, axis=1)
    y_pred_class = np.argmax(y_pred, axis=1)

    accuracy = accuracy_score(y_true_class, y_pred_class)
    precision = precision_score(y_true_class, y_pred_class, average='weighted')
    recall = recall_score(y_true_class, y_pred_class, average='weighted')
    f1 = f1_score(y_true_class, y_pred_class, average='weighted')

    print("Model Performance Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    print("\nClassification Report:")
    print(classification_report(y_true_class, y_pred_class))

    cm = confusion_matrix(y_true_class, y_pred_class)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

def train_model(X_train, y_train, X_val, y_val):
    input_shape = (X_train.shape[1], X_train.shape[2])
    model = create_model(input_shape)

    early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_loss', mode='min')

    history = model.fit(
        X_train, y_train,
        epochs=200,
        batch_size=32,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping, model_checkpoint],
        verbose=1
    )

    # Plot training history
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    return model

def out_of_sample_test(data, train_end, test_end):
    print("Performing out-of-sample test...")

    train_data = data[:train_end]
    test_data = data[train_end:test_end]

    print(f"Train data shape: {train_data.shape}")
    print(f"Test data shape: {test_data.shape}")

    train_features = create_features(train_data)
    test_features = create_features(test_data)

    X_train, y_train, scaler = prepare_data(train_features)
    if X_train is None or y_train is None:
        print("No valid training data after preprocessing")
        return None, None

    X_test, y_test, _ = prepare_data(test_features, scaler=scaler)
    if X_test is None or y_test is None:
        print("No valid testing data after preprocessing")
        return None, None

    print(f"Final training data shape: X_train {X_train.shape}, y_train {y_train.shape}")
    print(f"Final testing data shape: X_test {X_test.shape}, y_test {y_test.shape}")

    model = create_model(input_shape=(X_train.shape[1], X_train.shape[2]))

    print(model.summary())

    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_loss', mode='min')

    try:
        history = model.fit(
            X_train, y_train,
            epochs=100,
            batch_size=32,
            validation_split=0.2,
            callbacks=[early_stopping, model_checkpoint],
            verbose=1
        )

        # Plot training history
        plt.figure(figsize=(12, 6))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()

        # Evaluate the model
        test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
        print(f"Test Loss: {test_loss:.4f}")
        print(f"Test Accuracy: {test_accuracy:.4f}")

        return model, scaler

    except Exception as e:
        print(f"An error occurred during model training: {str(e)}")
        return None, None

def evaluate_predictions(y_true, y_pred):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    y_pred_binary = (y_pred > 0.5).astype(int)

    accuracy = accuracy_score(y_true, y_pred_binary)
    precision = precision_score(y_true, y_pred_binary, zero_division=0)
    recall = recall_score(y_true, y_pred_binary, zero_division=0)
    f1 = f1_score(y_true, y_pred_binary, zero_division=0)

    logging.info(f"Accuracy: {accuracy:.4f}")
    logging.info(f"Precision: {precision:.4f}")
    logging.info(f"Recall: {recall:.4f}")
    logging.info(f"F1 Score: {f1:.4f}")

def forward_test(data, model, scaler, start_date, end_date, window_size):
    logging.info("Performing forward test...")

    results = []
    dates = pd.date_range(start_date, end_date)

    for i in range(0, len(dates) - window_size, window_size):
        test_start = dates[i]
        test_end = dates[i + window_size]

        test_data = data[test_start:test_end]
        if len(test_data) == 0:
            logging.info(f"No data available for period {test_start} to {test_end}")
            continue

        test_features = create_features(test_data)
        X_test, y_test, _ = prepare_data(test_features, scaler=scaler)

        if X_test is None or y_test is None:
            logging.info(f"No valid features for period {test_start} to {test_end}")
            continue

        y_pred = model.predict(X_test)

        evaluate_predictions(y_test, y_pred)

        results.append({
            'start_date': test_start,
            'end_date': test_end,
            'accuracy': accuracy_score(y_test, (y_pred > 0.5).astype(int)),
            'precision': precision_score(y_test, (y_pred > 0.5).astype(int), zero_division=0),
            'recall': recall_score(y_test, (y_pred > 0.5).astype(int), zero_division=0),
            'f1': f1_score(y_test, (y_pred > 0.5).astype(int), zero_division=0)
        })

    if not results:
        logging.warning("No valid results for forward test. Check your date range and data.")
        return

    results_df = pd.DataFrame(results)
    logging.info(results_df.describe())

    plt.figure(figsize=(12, 6))
    plt.plot(results_df['start_date'], results_df['accuracy'], label='Accuracy')
    plt.plot(results_df['start_date'], results_df['f1'], label='F1 Score')
    plt.title('Model Performance Over Time')
    plt.xlabel('Date')
    plt.ylabel('Score')
    plt.legend()
    plt.show()

# Main execution
logging.info("Starting main execution...")
data = pd.read_csv('/content/hourly_SPY_2008-2021.csv', index_col='date', parse_dates=True)
logging.info(f"Loaded data shape: {data.shape}")

train_end = '2019-12-31'
val_end = '2020-12-31'
test_end = data.index.max()
logging.info(f"Train end: {train_end}")
logging.info(f"Validation end: {val_end}")
logging.info(f"Test end: {test_end}")

train_data = data[:train_end]
val_data = data[train_end:val_end]
test_data = data[val_end:test_end]

train_features = create_features(train_data)
val_features = create_features(val_data)
test_features = create_features(test_data)

X_train, y_train, scaler = prepare_data(train_features)
X_val, y_val, _ = prepare_data(val_features, scaler=scaler)
X_test, y_test, _ = prepare_data(test_features, scaler=scaler)

if X_train is not None and y_train is not None and X_val is not None and y_val is not None:
    model = train_model(X_train, y_train, X_val, y_val)

    logging.info("Evaluating on test set:")
    test_predictions = model.predict(X_test)
    evaluate_predictions(y_test, test_predictions)

    forward_start = pd.to_datetime('2021-01-01')
    forward_end = data.index.max()
    window_size = 7
    logging.info(f"Forward test range: {forward_start} to {forward_end}")
    forward_test(data, model, scaler, forward_start, forward_end, window_size)
else:
    logging.error("Failed to create valid training or validation data")
```
