```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             classification_report, confusion_matrix, roc_auc_score,
                             balanced_accuracy_score, roc_curve)
from sklearn.model_selection import TimeSeriesSplit
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf

# If you're using Keras, import it separately
from tensorflow import keras
from tensorflow.keras import backend as K
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from scipy.stats import gmean
import pickle
import requests


# Other imports
import talib
import logging
from sklearn.metrics import precision_recall_curve
from imblearn.over_sampling import SMOTE
import sys
import jax
import time

print(f"Python version: {sys.version}")
print(f"TensorFlow version: {tf.__version__}")
print(f"JAX version: {jax.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))

# Alpha Vantage API key
ALPHA_VANTAGE_API_KEY = '70R97N884LL7QPFU'

def fetch_spy_daily_data(api_key, output_file='SPY_daily_data.csv', max_retries=3):
    """Fetch SPY daily data from Alpha Vantage and save to CSV."""
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=SPY&apikey={api_key}'

    for attempt in range(max_retries):
        try:
            r = requests.get(url)
            r.raise_for_status()  # Raise an exception for bad status codes
            data = r.json()

            if 'Time Series (Daily)' not in data:
                if 'Note' in data:
                    print(f"API limit reached: {data['Note']}")
                    print("Waiting for 60 seconds before retrying...")
                    time.sleep(60)
                    continue
                else:
                    print("Unexpected data format received from Alpha Vantage.")
                    print("Response content:", data)
                    return None

            df = pd.DataFrame(data['Time Series (Daily)']).T
            df.index = pd.to_datetime(df.index)
            df = df.sort_index()
            df = df.rename(columns={
                '1. open': 'open',
                '2. high': 'high',
                '3. low': 'low',
                '4. close': 'close',
                '5. volume': 'volume'
            })
            df = df.astype(float)

            df.to_csv(output_file)
            print(f"Daily SPY data saved to {output_file}")
            return df

        except requests.exceptions.RequestException as e:
            print(f"Error fetching data from Alpha Vantage (Attempt {attempt + 1}/{max_retries}):")
            print(str(e))
            if attempt < max_retries - 1:
                print("Retrying in 5 seconds...")
                time.sleep(5)
            else:
                print("Max retries reached. Please check your internet connection and API key.")
                return None

    return None

def create_enhanced_features_daily(data):
    features = pd.DataFrame(index=data.index)

    # Basic return features
    features['returns'] = data['close'].pct_change()
    features['log_return'] = np.log(data['close'] / data['close'].shift(1))

    # Enhanced return dynamics
    features['return_acceleration'] = features['returns'].diff()
    features['log_return_5'] = np.log(data['close'] / data['close'].shift(5))
    features['return_5'] = data['close'].pct_change(5)

    # Trend and momentum indicators
    features['adx'] = talib.ADX(data['high'], data['low'], data['close'], timeperiod=14)
    features['rsi'] = talib.RSI(data['close'], timeperiod=14)
    features['rsi_momentum'] = features['rsi'].diff()

    # Volatility
    features['bb_width'] = (talib.BBANDS(data['close'], timeperiod=20)[0] - talib.BBANDS(data['close'], timeperiod=20)[2]) / talib.BBANDS(data['close'], timeperiod=20)[1]
    features['atr'] = talib.ATR(data['high'], data['low'], data['close'], timeperiod=14)

    # Non-linear interactions
    features['returns_adx'] = features['returns'] * features['adx']
    features['rsi_volatility'] = features['rsi'] * features['bb_width']

    # Volume-based features
    features['obv'] = talib.OBV(data['close'], data['volume'])
    features['volume_change'] = data['volume'].pct_change()

    # Keep only the most important features based on previous analysis
    important_features = ['returns', 'log_return', 'adx', 'rsi', 'rsi_momentum', 'bb_width',
                          'return_acceleration', 'log_return_5', 'return_5', 'atr',
                          'returns_adx', 'rsi_volatility', 'obv', 'volume_change']

    features = features[important_features]

    # Replace infinite values with NaN and then fill NaN values
    features = features.replace([np.inf, -np.inf], np.nan)
    features = features.fillna(method='ffill').fillna(method='bfill')

    return features

def prepare_data_daily(features, sequence_length=60, threshold=0.001):
    y = (features['returns'].shift(-1) > threshold).astype(int)
    features = features[:-1]
    y = y[:-1]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(features)

    X_seq, y_seq = [], []
    for i in range(len(X_scaled) - sequence_length):
        X_seq.append(X_scaled[i:i+sequence_length])
        y_seq.append(y.iloc[i+sequence_length])

    return np.array(X_seq), np.array(y_seq), scaler

def backtest_daily(model, X, data, confidence_threshold=0.3, stop_loss=0.03, take_profit=0.05,
                   max_position_size=1.0, min_position_size=0.1, drawdown_limit=0.2):
    predictions = model.predict(X).flatten()

    position = 0
    entry_price = 0
    entry_time = 0
    trades = []
    equity = [1.0]
    drawdowns = [0.0]
    holding_times = []
    recent_trades = []

    def calculate_drawdown(equity):
        peak = max(equity)
        return (peak - equity[-1]) / peak

    def get_dynamic_position_size(confidence, drawdown, recent_performance):
        if confidence < confidence_threshold:
            return 0

        drawdown_factor = max(0, 1 - (drawdown / drawdown_limit))

        performance_factor = 1.0
        if recent_trades:
            win_rate = sum(1 for trade in recent_trades if trade > 0) / len(recent_trades)
            performance_factor = 0.5 + win_rate

        base_size = max_position_size * drawdown_factor * performance_factor

        confidence_factor = (confidence - confidence_threshold) / (1 - confidence_threshold)
        dynamic_size = base_size * (0.5 + 0.5 * confidence_factor)

        return max(min(dynamic_size, max_position_size), min_position_size)

    for i in range(len(predictions)):
        current_price = data['close'].iloc[i]
        current_time = data.index[i]

        confidence = predictions[i]
        current_drawdown = calculate_drawdown(equity)
        position_size = get_dynamic_position_size(confidence, current_drawdown, recent_trades)

        if current_drawdown > drawdown_limit:
            position_size = 0

        if position == 0 and position_size > 0:
            entry_price = current_price
            position = position_size
            entry_time = i
            trades.append(('buy', i, entry_price, position_size))
        elif position > 0:
            exit_price = current_price
            returns = (exit_price - entry_price) / entry_price
            if returns <= -stop_loss or returns >= take_profit or current_drawdown > drawdown_limit:
                trades.append(('sell', i, exit_price, position))
                trade_return = position * returns
                equity.append(equity[-1] * (1 + trade_return))
                holding_times.append(i - entry_time)
                recent_trades.append(trade_return)
                if len(recent_trades) > 10:
                    recent_trades.pop(0)
                position = 0

        if len(equity) > 1:
            drawdown = calculate_drawdown(equity)
            drawdowns.append(drawdown)

    if position > 0:
        exit_price = data['close'].iloc[-1]
        returns = (exit_price - entry_price) / entry_price
        trades.append(('sell', len(predictions)-1, exit_price, position))
        equity.append(equity[-1] * (1 + position * returns))
        holding_times.append(len(predictions) - 1 - entry_time)

    total_trades = len(trades) // 2
    winning_trades = sum(1 for i in range(0, len(trades), 2) if trades[i+1][2] > trades[i][2])
    losing_trades = total_trades - winning_trades
    win_rate = winning_trades / total_trades if total_trades > 0 else 0

    total_return = equity[-1] - 1
    sharpe_ratio = np.mean(np.diff(equity)) / np.std(np.diff(equity)) * np.sqrt(252) if len(equity) > 1 else 0
    max_drawdown = max(drawdowns)
    avg_holding_time = np.mean(holding_times) if holding_times else 0

    # Calculate CAGR
    start_date = data.index[0]
    end_date = data.index[-1]
    years = (end_date - start_date).days / 365.25
    cagr = (equity[-1] ** (1/years) - 1) if years > 0 else 0

    # Calculate average trades per year
    avg_trades_per_year = total_trades / years if years > 0 else 0

    print(f"Confidence Threshold: {confidence_threshold:.2f}")
    print(f"Total Trades: {total_trades}")
    print(f"Winning Trades: {winning_trades}")
    print(f"Losing Trades: {losing_trades}")
    print(f"Win Rate: {win_rate:.2%}")
    print(f"Total Return: {total_return:.2%}")
    print(f"CAGR: {cagr:.2%}")
    print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
    print(f"Max Drawdown: {max_drawdown:.2%}")
    print(f"Average Holding Time: {avg_holding_time:.2f} days")
    print(f"Average Trades per Year: {avg_trades_per_year:.2f}")

    plt.figure(figsize=(12, 6))
    plt.plot(equity)
    plt.title(f'Equity Curve (Threshold: {confidence_threshold:.2f})')
    plt.xlabel('Trade Number')
    plt.ylabel('Equity')
    plt.show()

    return {
        'confidence_threshold': confidence_threshold,
        'stop_loss': stop_loss,
        'take_profit': take_profit,
        'total_trades': total_trades,
        'win_rate': win_rate,
        'total_return': total_return,
        'cagr': cagr,
        'sharpe_ratio': sharpe_ratio,
        'max_drawdown': max_drawdown,
        'avg_holding_time': avg_holding_time,
        'avg_trades_per_year': avg_trades_per_year
    }

def optimize_sl_tp_daily(model, X, data, confidence_threshold=0.3):
    sl_values = [0.03, 0.05, 0.07]
    tp_values = [0.05, 0.07, 0.09]

    results = []

    for sl in sl_values:
        for tp in tp_values:
            print(f"\nTesting SL: {sl:.2f}, TP: {tp:.2f}")
            result = backtest_daily(model, X, data,
                                    confidence_threshold=confidence_threshold,
                                    stop_loss=sl,
                                    take_profit=tp,
                                    max_position_size=1.0,
                                    min_position_size=0.1,
                                    drawdown_limit=0.2)
            results.append(result)

            print(f"Sharpe Ratio: {result['sharpe_ratio']:.2f}")
            print(f"Total Return: {result['total_return']:.2%}")
            print(f"CAGR: {result['cagr']:.2%}")
            print(f"Max Drawdown: {result['max_drawdown']:.2%}")
            print(f"Win Rate: {result['win_rate']:.2%}")
            print(f"Avg Trades per Year: {result['avg_trades_per_year']:.2f}")
            print("--------------------")

    # Find the best result based on Sharpe ratio
    best_result = max(results, key=lambda x: x['sharpe_ratio'])

    print("\nBest Result:")
    print(f"SL: {best_result['stop_loss']:.2f}, TP: {best_result['take_profit']:.2f}")
    print(f"Sharpe Ratio: {best_result['sharpe_ratio']:.2f}")
    print(f"Total Return: {best_result['total_return']:.2%}")
    print(f"CAGR: {best_result['cagr']:.2%}")
    print(f"Max Drawdown: {best_result['max_drawdown']:.2%}")
    print(f"Win Rate: {best_result['win_rate']:.2%}")
    print(f"Avg Trades per Year: {best_result['avg_trades_per_year']:.2f}")

    return results, best_result

def load_spy_data(file_path):
    """Load SPY daily data from CSV file."""
    df = pd.read_csv(file_path, parse_dates=['date'], index_col='date')
    df.columns = [col.strip() for col in df.columns]  # Remove any leading/trailing whitespace from column names
    print(f"Data loaded. Shape: {df.shape}")
    print(f"Date range: {df.index.min()} to {df.index.max()}")
    print("Columns:", df.columns.tolist())
    return df

def create_hybrid_model(input_shape, num_heads=4, ff_dim=64, num_transformer_blocks=4, mlp_units=[64, 32], dropout=0.3):
    inputs = Input(shape=input_shape)

    x = Bidirectional(LSTM(64, return_sequences=True))(inputs)
    x = Bidirectional(LSTM(64, return_sequences=True))(x)

    # Adjust LSTM output to match input feature dimension
    x = Dense(input_shape[-1])(x)

    for _ in range(num_transformer_blocks):
        attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=ff_dim)(x, x)
        x = LayerNormalization(epsilon=1e-6)(x + attn_output)
        ff_output = Dense(ff_dim, activation="relu")(x)
        ff_output = Dense(input_shape[-1])(ff_output)
        x = LayerNormalization(epsilon=1e-6)(x + ff_output)

    x = GlobalAveragePooling1D()(x)
    for dim in mlp_units:
        x = Dense(dim, activation="relu")(x)
        x = Dropout(dropout)(x)
    outputs = Dense(1, activation="sigmoid")(x)

    return Model(inputs=inputs, outputs=outputs)

def focal_loss(gamma=2., alpha=.25):
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)

        pt_1 = tf.where(tf.equal(y_true, 1), y_pred, tf.ones_like(y_pred))
        pt_0 = tf.where(tf.equal(y_true, 0), y_pred, tf.zeros_like(y_pred))

        return -tf.reduce_mean(alpha * tf.pow(1. - pt_1, gamma) * tf.math.log(pt_1 + epsilon)) - \
               tf.reduce_mean((1 - alpha) * tf.pow(pt_0, gamma) * tf.math.log(1. - pt_0 + epsilon))
    return focal_loss_fixed

def aggregate_to_5min(data, output_filename='SPY_data_5_min.csv'):
    """Aggregate 1-minute data to 5-minute intervals, removing rows with missing OHLC data, and save to CSV."""
    logging.info("Aggregating 1-minute data to 5-minute intervals...")

    # Remove rows where any OHLC data is missing
    data = data.dropna(subset=['open', 'high', 'low', 'close'])

    # Ensure the index is datetime
    data.index = pd.to_datetime(data.index)

    agg_dict = {
        'open': 'first',
        'high': 'max',
        'low': 'min',
        'close': 'last',
        'volume': 'sum'
    }

    data_5min = data.resample('5T').agg(agg_dict)

    # Remove any aggregated rows with NaN values
    data_5min = data_5min.dropna()

    # Save the aggregated data to a CSV file
    data_5min.to_csv(output_filename)

    logging.info(f"Original data shape: {data.shape}")
    logging.info(f"Aggregated data shape: {data_5min.shape}")
    logging.info(f"Aggregated data saved to {output_filename}")

    return data_5min

def feature_importance_3d(model, X, y, feature_names):
    n_features = X.shape[2]
    importances = np.zeros(n_features)

    # Compute baseline score
    baseline_score = model.evaluate(X, y, verbose=0)[1]  # Use accuracy as the metric

    for i in range(n_features):
        # Create a copy of the input data
        X_permuted = X.copy()

        # Permute the i-th feature across all sequences
        X_permuted[:, :, i] = np.random.permutation(X_permuted[:, :, i])

        # Compute new score
        new_score = model.evaluate(X_permuted, y, verbose=0)[1]

        # Compute importance
        importances[i] = baseline_score - new_score

    # Create DataFrame with feature importances
    importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
    importance_df = importance_df.sort_values('importance', ascending=False)

    # Plot feature importances
    plt.figure(figsize=(10, 6))
    sns.barplot(x='importance', y='feature', data=importance_df.head(20))
    plt.title('Feature Importance')
    plt.tight_layout()
    plt.show()

def create_features_daily(data):
    features = pd.DataFrame(index=data.index)

    # Basic return features
    features['returns'] = data['Close'].pct_change()
    features['log_return'] = np.log(data['Close'] / data['Close'].shift(1))

    # Enhanced return dynamics
    features['return_acceleration'] = features['returns'].diff()
    features['log_return_5'] = np.log(data['Close'] / data['Close'].shift(5))
    features['return_5'] = data['Close'].pct_change(5)

    # Trend and momentum indicators
    features['adx'] = talib.ADX(data['High'], data['Low'], data['Close'])
    features['rsi'] = talib.RSI(data['Close'])
    features['rsi_momentum'] = features['rsi'].diff()

    # Volatility
    features['bb_width'] = (talib.BBANDS(data['Close'])[0] - talib.BBANDS(data['Close'])[2]) / talib.BBANDS(data['Close'])[1]
    features['atr'] = talib.ATR(data['High'], data['Low'], data['Close'])

    # Non-linear interactions
    features['returns_adx'] = features['returns'] * features['adx']
    features['rsi_volatility'] = features['rsi'] * features['bb_width']

    # Volume-based features
    features['obv'] = talib.OBV(data['Close'], data['Volume'])
    features['volume_change'] = data['Volume'].pct_change()

    # Replace infinite values with NaN and then fill NaN values
    features = features.replace([np.inf, -np.inf], np.nan)
    features = features.fillna(method='ffill').fillna(method='bfill')

    return features

def create_model_daily(input_shape):
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=input_shape),
        LSTM(32),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

def prepare_data(features, sequence_length=60, threshold=0.001):
    y = (features['returns'].shift(-1) > threshold).astype(int)
    features = features[:-1]
    y = y[:-1]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(features)

    X_seq, y_seq = [], []
    for i in range(len(X_scaled) - sequence_length):
        X_seq.append(X_scaled[i:i+sequence_length])
        y_seq.append(y.iloc[i+sequence_length])

    return np.array(X_seq), np.array(y_seq), scaler

def calculate_cagr(start_value, end_value, num_years):
    return (end_value / start_value) ** (1 / num_years) - 1

def backtest_daily(model, X, data, confidence_threshold=0.3, stop_loss=0.03, take_profit=0.05,
                   max_position_size=1.0, min_position_size=0.1, drawdown_limit=0.2):
    predictions = model.predict(X).flatten()

    position = 0
    entry_price = 0
    entry_time = 0
    trades = []
    equity = [1.0]
    drawdowns = [0.0]
    holding_times = []
    recent_trades = []

    def calculate_drawdown(equity):
        peak = max(equity)
        return (peak - equity[-1]) / peak

    def get_dynamic_position_size(confidence, drawdown, recent_performance):
        if confidence < confidence_threshold:
            return 0

        drawdown_factor = max(0, 1 - (drawdown / drawdown_limit))

        performance_factor = 1.0
        if recent_trades:
            win_rate = sum(1 for trade in recent_trades if trade > 0) / len(recent_trades)
            performance_factor = 0.5 + win_rate

        base_size = max_position_size * drawdown_factor * performance_factor

        confidence_factor = (confidence - confidence_threshold) / (1 - confidence_threshold)
        dynamic_size = base_size * (0.5 + 0.5 * confidence_factor)

        return max(min(dynamic_size, max_position_size), min_position_size)

    for i in range(len(predictions)):
        current_price = data['close'].iloc[i]
        current_time = data.index[i]

        confidence = predictions[i]
        current_drawdown = calculate_drawdown(equity)
        position_size = get_dynamic_position_size(confidence, current_drawdown, recent_trades)

        if current_drawdown > drawdown_limit:
            position_size = 0

        if position == 0 and position_size > 0:
            entry_price = current_price
            position = position_size
            entry_time = i
            trades.append(('buy', i, entry_price, position_size))
        elif position > 0:
            exit_price = current_price
            returns = (exit_price - entry_price) / entry_price
            if returns <= -stop_loss or returns >= take_profit or current_drawdown > drawdown_limit:
                trades.append(('sell', i, exit_price, position))
                trade_return = position * returns
                equity.append(equity[-1] * (1 + trade_return))
                holding_times.append(i - entry_time)
                recent_trades.append(trade_return)
                if len(recent_trades) > 10:
                    recent_trades.pop(0)
                position = 0

        if len(equity) > 1:
            drawdown = calculate_drawdown(equity)
            drawdowns.append(drawdown)

    if position > 0:
        exit_price = data['close'].iloc[-1]
        returns = (exit_price - entry_price) / entry_price
        trades.append(('sell', len(predictions)-1, exit_price, position))
        equity.append(equity[-1] * (1 + position * returns))
        holding_times.append(len(predictions) - 1 - entry_time)

    total_trades = len(trades) // 2
    winning_trades = sum(1 for i in range(0, len(trades), 2) if trades[i+1][2] > trades[i][2])
    losing_trades = total_trades - winning_trades
    win_rate = winning_trades / total_trades if total_trades > 0 else 0

    total_return = equity[-1] - 1
    sharpe_ratio = np.mean(np.diff(equity)) / np.std(np.diff(equity)) * np.sqrt(252) if len(equity) > 1 else 0
    max_drawdown = max(drawdowns)
    avg_holding_time = np.mean(holding_times) if holding_times else 0

    # Calculate CAGR
    start_date = data.index[0]
    end_date = data.index[-1]
    years = (end_date - start_date).days / 365.25
    cagr = (equity[-1] ** (1/years) - 1) if years > 0 else 0

    # Calculate average trades per year
    avg_trades_per_year = total_trades / years if years > 0 else 0

    print(f"Confidence Threshold: {confidence_threshold:.2f}")
    print(f"Total Trades: {total_trades}")
    print(f"Winning Trades: {winning_trades}")
    print(f"Losing Trades: {losing_trades}")
    print(f"Win Rate: {win_rate:.2%}")
    print(f"Total Return: {total_return:.2%}")
    print(f"CAGR: {cagr:.2%}")
    print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
    print(f"Max Drawdown: {max_drawdown:.2%}")
    print(f"Average Holding Time: {avg_holding_time:.2f} days")
    print(f"Average Trades per Year: {avg_trades_per_year:.2f}")

    plt.figure(figsize=(12, 6))
    plt.plot(equity)
    plt.title(f'Equity Curve (Threshold: {confidence_threshold:.2f})')
    plt.xlabel('Trade Number')
    plt.ylabel('Equity')
    plt.show()

    return {
        'confidence_threshold': confidence_threshold,
        'stop_loss': stop_loss,
        'take_profit': take_profit,
        'total_trades': total_trades,
        'win_rate': win_rate,
        'total_return': total_return,
        'cagr': cagr,
        'sharpe_ratio': sharpe_ratio,
        'max_drawdown': max_drawdown,
        'avg_holding_time': avg_holding_time,
        'avg_trades_per_year': avg_trades_per_year
    }

def subperiod_analysis_daily(model, data, features, sequence_length=60):
    subperiods = [
        ('Full Period', data.index[0], data.index[-1]),
        ('Global Financial Crisis', '2008-01-01', '2009-06-30'),
        ('Post-Crisis Recovery', '2009-07-01', '2013-12-31'),
        ('Bull Market', '2014-01-01', '2019-12-31'),
        ('Covid-19 Crash', '2020-01-01', '2020-03-31'),
        ('Post-Covid Recovery', '2020-04-01', data.index[-1])
    ]

    results = {}

    for period_name, start_date, end_date in subperiods:
        print(f"\nAnalyzing {period_name}: {start_date} to {end_date}")

        start_date = pd.to_datetime(start_date)
        end_date = pd.to_datetime(end_date)

        mask = (data.index >= start_date) & (data.index <= end_date)
        subperiod_data = data[mask]
        subperiod_features = features[mask]

        if len(subperiod_data) <= sequence_length:
            print(f"Insufficient data for {period_name}. Skipping this period.")
            continue

        X_sub, _ = prepare_data_daily(subperiod_features, sequence_length)

        result = backtest_daily(model, X_sub, subperiod_data)
        results[period_name] = result

        print(f"Total Return: {result['total_return']:.2%}")
        print(f"CAGR: {result['cagr']:.2%}")
        print(f"Sharpe Ratio: {result['sharpe_ratio']:.2f}")
        print(f"Max Drawdown: {result['max_drawdown']:.2%}")
        print(f"Win Rate: {result['win_rate']:.2%}")

    # Plotting
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle('NIFTY 100 Strategy Performance Across Different Market Periods', fontsize=16)
    metrics = ['total_return', 'sharpe_ratio', 'max_drawdown', 'win_rate', 'total_trades', 'avg_holding_time']
    titles = ['Total Return', 'Sharpe Ratio', 'Max Drawdown', 'Win Rate', 'Total Trades', 'Avg Holding Time (Days)']

    for i, (metric, title) in enumerate(zip(metrics, titles)):
        ax = axes[i // 3, i % 3]
        values = [results[period][metric] for period in results]
        ax.bar(results.keys(), values)
        ax.set_title(title)
        ax.set_xticklabels(results.keys(), rotation=45, ha='right')
        if metric in ['total_return', 'max_drawdown', 'win_rate']:
            ax.set_yticklabels([f'{x:.0%}' for x in ax.get_yticks()])

    plt.tight_layout()
    plt.show()

    return results

def subperiod_analysis_daily(model, X, data, scaler):
    subperiods = [
        ('Full Period', data.index[0], data.index[-1]),
        ('Global Financial Crisis', '2008-01-01', '2009-06-30'),
        ('Post-Crisis Recovery', '2009-07-01', '2013-12-31'),
        ('Bull Market', '2014-01-01', '2019-12-31'),
        ('Covid-19 Crash', '2020-01-01', '2020-03-31'),
        ('Post-Covid Recovery', '2020-04-01', data.index[-1])
    ]

    results = {}

    for period_name, start_date, end_date in subperiods:
        print(f"\nAnalyzing {period_name}: {start_date} to {end_date}")

        mask = (data.index >= start_date) & (data.index <= end_date)
        subperiod_data = data[mask]

        subperiod_features = create_features_daily(subperiod_data)
        X_sub = prepare_data_daily(subperiod_features, scaler)

        result = backtest_daily(model, X_sub, subperiod_data)
        results[period_name] = result

        print(f"Total Return: {result['total_return']:.2%}")
        print(f"Sharpe Ratio: {result['sharpe_ratio']:.2f}")
        print(f"Max Drawdown: {result['max_drawdown']:.2%}")
        print(f"Win Rate: {result['win_rate']:.2%}")

    # Plotting
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle('NIFTY 100 Strategy Performance Across Different Market Periods', fontsize=16)
    metrics = ['total_return', 'sharpe_ratio', 'max_drawdown', 'win_rate', 'total_trades', 'avg_holding_time']
    titles = ['Total Return', 'Sharpe Ratio', 'Max Drawdown', 'Win Rate', 'Total Trades', 'Avg Holding Time (Days)']

    for i, (metric, title) in enumerate(zip(metrics, titles)):
        ax = axes[i // 3, i % 3]
        values = [results[period][metric] for period in results]
        ax.bar(results.keys(), values)
        ax.set_title(title)
        ax.set_xticklabels(results.keys(), rotation=45, ha='right')
        if metric in ['total_return', 'max_drawdown', 'win_rate']:
            ax.set_yticklabels([f'{x:.0%}' for x in ax.get_yticks()])

    plt.tight_layout()
    plt.show()

    return results

def create_enhanced_features(data):
    features = pd.DataFrame(index=data.index)

    # Basic return features
    features['returns'] = data['4. close'].pct_change()
    features['log_return'] = np.log(data['4. close'] / data['4. close'].shift(1))

    # Enhanced return dynamics
    features['return_acceleration'] = features['returns'].diff()
    features['log_return_5'] = np.log(data['4. close'] / data['4. close'].shift(5))
    features['return_5'] = data['4. close'].pct_change(5)

    # Trend and momentum indicators
    features['adx'] = talib.ADX(data['2. high'], data['3. low'], data['4. close'])
    features['rsi'] = talib.RSI(data['4. close'])
    features['rsi_momentum'] = features['rsi'].diff()

    # Volatility
    features['bb_width'] = (talib.BBANDS(data['4. close'])[0] - talib.BBANDS(data['4. close'])[2]) / talib.BBANDS(data['4. close'])[1]
    features['atr'] = talib.ATR(data['2. high'], data['3. low'], data['4. close'])

    # Non-linear interactions
    features['returns_adx'] = features['returns'] * features['adx']
    features['rsi_volatility'] = features['rsi'] * features['bb_width']

    # Volume-based features
    features['obv'] = talib.OBV(data['4. close'], data['5. volume'])
    features['volume_change'] = data['5. volume'].pct_change()

    # Keep only the most important features based on previous analysis
    important_features = ['returns', 'log_return', 'adx', 'rsi', 'rsi_momentum', 'bb_width',
                          'return_acceleration', 'log_return_5', 'return_5', 'atr',
                          'returns_adx', 'rsi_volatility', 'obv', 'volume_change']

    features = features[important_features]

    # Replace infinite values with NaN and then fill NaN values
    features = features.replace([np.inf, -np.inf], np.nan)
    features = features.fillna(method='ffill').fillna(method='bfill')

    return features


def prepare_data(features, sequence_length=60, threshold=0.001):
    y = (features['returns'].shift(-1) > threshold).astype(int)
    features = features[:-1]
    y = y[:-1]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(features)

    X_seq, y_seq = [], []
    for i in range(len(X_scaled) - sequence_length):
        X_seq.append(X_scaled[i:i+sequence_length])
        y_seq.append(y.iloc[i+sequence_length])

    return np.array(X_seq), np.array(y_seq), scaler

def create_transformer_model(input_shape, num_heads=4, ff_dim=64, num_transformer_blocks=4, mlp_units=[64, 32], dropout=0.3):
    inputs = Input(shape=input_shape)
    x = inputs
    for _ in range(num_transformer_blocks):
        attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=ff_dim)(x, x)
        attn_output = Dense(input_shape[-1])(attn_output)
        attn_output = Dropout(dropout)(attn_output)
        x = LayerNormalization(epsilon=1e-6)(x + attn_output)
        ff_output = Dense(ff_dim, activation="relu")(x)
        ff_output = Dense(input_shape[-1])(ff_output)
        ff_output = Dropout(dropout)(ff_output)
        x = LayerNormalization(epsilon=1e-6)(x + ff_output)

    x = GlobalAveragePooling1D()(x)
    for dim in mlp_units:
        x = Dense(dim, activation="relu")(x)
        x = Dropout(dropout)(x)
    outputs = Dense(1, activation="sigmoid")(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

def train_model(X_train, y_train, X_val, y_val, model_path='best_model.h5'):
    model = create_hybrid_model(input_shape=(X_train.shape[1], X_train.shape[2]))

    model.compile(optimizer=Adam(learning_rate=1e-4),
                  loss=focal_loss(alpha=0.75, gamma=2),
                  metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint(model_path, save_best_only=True, monitor='val_loss', mode='min')
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)

    history = model.fit(
        X_train, y_train,
        epochs=100,
        batch_size=64,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping, model_checkpoint, reduce_lr],
        class_weight={0: 1, 1: 10},
        verbose=1
    )

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f"Model saved to {model_path}")
    return model, history

def optimize_sl_tp(model, X, data, confidence_threshold=0.3):
    sl_values = [0.03]  # Keep stop-loss constant at 0.03
    tp_values = [0.09, 0.10, 0.11, 0.12]  # New take-profit values

    results = []

    for sl in sl_values:
        for tp in tp_values:
            print(f"\nTesting SL: {sl:.2f}, TP: {tp:.2f}")
            result = backtest(model, X, data,
                              confidence_threshold=confidence_threshold,
                              stop_loss=sl,
                              take_profit=tp,
                              max_position_size=1.0,
                              min_position_size=0.1,
                              drawdown_limit=0.2)
            results.append(result)

            print(f"Sharpe Ratio: {result['sharpe_ratio']:.2f}")
            print(f"Total Return: {result['total_return']:.2%}")
            print(f"CAGR: {result['cagr']:.2%}")
            print(f"Max Drawdown: {result['max_drawdown']:.2%}")
            print(f"Win Rate: {result['win_rate']:.2%}")
            print(f"Avg Trades per Year: {result['avg_trades_per_year']:.2f}")
            print("--------------------")

    # Find the best result based on Sharpe ratio
    best_result = max(results, key=lambda x: x['sharpe_ratio'])

    print("\nBest Result:")
    print(f"SL: {best_result['stop_loss']:.2f}, TP: {best_result['take_profit']:.2f}")
    print(f"Sharpe Ratio: {best_result['sharpe_ratio']:.2f}")
    print(f"Total Return: {best_result['total_return']:.2%}")
    print(f"CAGR: {best_result['cagr']:.2%}")
    print(f"Max Drawdown: {best_result['max_drawdown']:.2%}")
    print(f"Win Rate: {best_result['win_rate']:.2%}")
    print(f"Avg Trades per Year: {best_result['avg_trades_per_year']:.2f}")

    return results, best_result


def backtest(model, X, data, confidence_threshold=0.3, stop_loss=0.03, take_profit=0.05,
             max_position_size=1.0, min_position_size=0.1, drawdown_limit=0.2):
    predictions = model.predict(X).flatten()

    position = 0
    entry_price = 0
    entry_time = 0
    trades = []
    equity = [1.0]
    drawdowns = [0.0]
    holding_times = []
    recent_trades = []

    def calculate_drawdown(equity):
        peak = max(equity)
        return (peak - equity[-1]) / peak

    def get_dynamic_position_size(confidence, drawdown, recent_performance):
        if confidence < confidence_threshold:
            return 0

        drawdown_factor = max(0, 1 - (drawdown / drawdown_limit))

        performance_factor = 1.0
        if recent_trades:
            win_rate = sum(1 for trade in recent_trades if trade > 0) / len(recent_trades)
            performance_factor = 0.5 + win_rate

        base_size = max_position_size * drawdown_factor * performance_factor

        confidence_factor = (confidence - confidence_threshold) / (1 - confidence_threshold)
        dynamic_size = base_size * (0.5 + 0.5 * confidence_factor)

        return max(min(dynamic_size, max_position_size), min_position_size)

    for i in range(len(predictions)):
        current_price = data['4. close'].iloc[i]
        current_time = data.index[i]

        confidence = predictions[i]
        current_drawdown = calculate_drawdown(equity)
        position_size = get_dynamic_position_size(confidence, current_drawdown, recent_trades)

        if current_drawdown > drawdown_limit:
            position_size = 0

        if position == 0 and position_size > 0:
            entry_price = current_price  # No slippage
            position = position_size
            entry_time = i
            trades.append(('buy', i, entry_price, position_size))
        elif position > 0:
            exit_price = current_price  # No slippage
            returns = (exit_price - entry_price) / entry_price
            if returns <= -stop_loss or returns >= take_profit or current_drawdown > drawdown_limit:
                trades.append(('sell', i, exit_price, position))
                trade_return = position * returns
                equity.append(equity[-1] * (1 + trade_return))  # No transaction costs
                holding_times.append(i - entry_time)
                recent_trades.append(trade_return)
                if len(recent_trades) > 10:
                    recent_trades.pop(0)
                position = 0

        if len(equity) > 1:
            drawdown = calculate_drawdown(equity)
            drawdowns.append(drawdown)

    if position > 0:
        exit_price = data['4. close'].iloc[-1]  # No slippage
        returns = (exit_price - entry_price) / entry_price
        trades.append(('sell', len(predictions)-1, exit_price, position))
        equity.append(equity[-1] * (1 + position * returns))  # No transaction costs
        holding_times.append(len(predictions) - 1 - entry_time)

    total_trades = len(trades) // 2
    winning_trades = sum(1 for i in range(0, len(trades), 2) if trades[i+1][2] > trades[i][2])
    losing_trades = total_trades - winning_trades
    win_rate = winning_trades / total_trades if total_trades > 0 else 0

    total_return = equity[-1] - 1
    sharpe_ratio = np.mean(np.diff(equity)) / np.std(np.diff(equity)) * np.sqrt(252) if len(equity) > 1 else 0
    max_drawdown = max(drawdowns)
    avg_holding_time = np.mean(holding_times) if holding_times else 0

    # Calculate CAGR
    start_date = data.index[0]
    end_date = data.index[-1]
    years = (end_date - start_date).days / 365.25
    cagr = (equity[-1] ** (1/years) - 1) if years > 0 else 0

    # Calculate average trades per year
    avg_trades_per_year = total_trades / years if years > 0 else 0

    print(f"Confidence Threshold: {confidence_threshold:.2f}")
    print(f"Total Trades: {total_trades}")
    print(f"Winning Trades: {winning_trades}")
    print(f"Losing Trades: {losing_trades}")
    print(f"Win Rate: {win_rate:.2%}")
    print(f"Total Return: {total_return:.2%}")
    print(f"CAGR: {cagr:.2%}")
    print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
    print(f"Max Drawdown: {max_drawdown:.2%}")
    print(f"Average Holding Time: {avg_holding_time:.2f} days")
    print(f"Average Trades per Year: {avg_trades_per_year:.2f}")

    plt.figure(figsize=(12, 6))
    plt.plot(equity)
    plt.title(f'Equity Curve (Threshold: {confidence_threshold:.2f})')
    plt.xlabel('Trade Number')
    plt.ylabel('Equity')
    plt.show()

    return {
        'confidence_threshold': confidence_threshold,
        'stop_loss': stop_loss,
        'take_profit': take_profit,
        'total_trades': total_trades,
        'win_rate': win_rate,
        'total_return': total_return,
        'cagr': cagr,
        'sharpe_ratio': sharpe_ratio,
        'max_drawdown': max_drawdown,
        'avg_holding_time': avg_holding_time,
        'avg_trades_per_year': avg_trades_per_year
    }

def feature_importance(model, X, y, feature_names):
    results = permutation_importance(model.predict, X, y, n_repeats=10, random_state=42)
    importance_df = pd.DataFrame({'feature': feature_names, 'importance': results.importances_mean})
    importance_df = importance_df.sort_values('importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='importance', y='feature', data=importance_df.head(20))
    plt.title('Feature Importance')
    plt.tight_layout()
    plt.show()

def plot_training_history(history):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_feature_importance(model, feature_names):
    # Get the weights of the first dense layer after the input
    weights = model.layers[1].get_weights()[0]

    # Calculate feature importance as the mean absolute value of weights
    importance = np.mean(np.abs(weights), axis=1)

    # Sort features by importance
    sorted_idx = np.argsort(importance)
    sorted_importance = importance[sorted_idx]
    sorted_features = np.array(feature_names)[sorted_idx]

    # Plot
    plt.figure(figsize=(10, 8))
    plt.barh(range(len(sorted_importance)), sorted_importance)
    plt.yticks(range(len(sorted_importance)), sorted_features)
    plt.xlabel('Importance')
    plt.title('Feature Importance')
    plt.tight_layout()
    plt.show()

def evaluate_model(model, X_test, y_test):
    y_pred_proba = model.predict(X_test)

    # Try different thresholds
    thresholds = [0.485, 0.495, 0.5, 0.505, 0.515]
    for threshold in thresholds:
        y_pred = (y_pred_proba > threshold).astype(int).flatten()
        balanced_acc = balanced_accuracy_score(y_test, y_pred)
        print(f"\nThreshold: {threshold:.2f}")
        print(f"Balanced Accuracy: {balanced_acc:.4f}")
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred))

    auc_roc = roc_auc_score(y_test, y_pred_proba)
    print(f"\nAUC-ROC: {auc_roc:.4f}")

    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc_roc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()

def is_last_interval(timestamp):
    return timestamp.time() == pd.Timestamp('14:10').time()

def get_position_size(confidence, threshold):
    if confidence >= 0.5:
        return 1.0
    elif confidence >= 0.45:
        return 0.75
    elif confidence >= 0.4:
        return 0.5
    elif confidence >= 0.35:
        return 0.25
    elif confidence >= threshold:
        return 0.1
    return 0

def subperiod_analysis_daily(model, data, features, sequence_length=60):
    subperiods = [
        ('Full Period', data.index[0], data.index[-1]),
        ('Global Financial Crisis', '2008-01-01', '2009-06-30'),
        ('Post-Crisis Recovery', '2009-07-01', '2013-12-31'),
        ('Bull Market', '2014-01-01', '2019-12-31'),
        ('Covid-19 Crash', '2020-01-01', '2020-03-31'),
        ('Post-Covid Recovery', '2020-04-01', data.index[-1])
    ]

    results = {}

    for period_name, start_date, end_date in subperiods:
        print(f"\nAnalyzing {period_name}: {start_date} to {end_date}")

        mask = (data.index >= start_date) & (data.index <= end_date)
        subperiod_data = data[mask]
        subperiod_features = features[mask]

        if len(subperiod_data) <= sequence_length:
            print(f"Insufficient data for {period_name}. Skipping this period.")
            continue

        X_sub, _ = prepare_data_daily(subperiod_features, sequence_length)

        result = backtest_daily(model, X_sub, subperiod_data)
        results[period_name] = result

        print(f"Total Return: {result['total_return']:.2%}")
        print(f"CAGR: {result['cagr']:.2%}")
        print(f"Sharpe Ratio: {result['sharpe_ratio']:.2f}")
        print(f"Max Drawdown: {result['max_drawdown']:.2%}")
        print(f"Win Rate: {result['win_rate']:.2%}")

    # Plotting
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle('NIFTY 100 Strategy Performance Across Different Market Periods', fontsize=16)
    metrics = ['total_return', 'cagr', 'sharpe_ratio', 'max_drawdown', 'win_rate', 'total_trades']
    titles = ['Total Return', 'CAGR', 'Sharpe Ratio', 'Max Drawdown', 'Win Rate', 'Total Trades']

    for i, (metric, title) in enumerate(zip(metrics, titles)):
        ax = axes[i // 3, i % 3]
        values = [results[period][metric] for period in results if period in results]
        ax.bar(results.keys(), values)
        ax.set_title(title)
        ax.set_xticklabels(results.keys(), rotation=45, ha='right')
        if metric in ['total_return', 'cagr', 'max_drawdown', 'win_rate']:
            ax.set_yticklabels([f'{x:.0%}' for x in ax.get_yticks()])

    plt.tight_layout()
    plt.show()

    return results


    return results

# Main execution
if __name__ == "__main__":
    # Load SPY daily data from CSV
    data = load_spy_data('/content/spy_data_with_indicators_extended.csv')  # Update this path to your uploaded file location

    # Prepare features
    features = create_enhanced_features(data)
    X, y, scaler = prepare_data(features)

    # Load the pre-trained model
    model_path = '/content/best_model.h5'
    try:
        custom_objects = {'focal_loss_fixed': focal_loss()}
        loaded_model = load_model(model_path, custom_objects=custom_objects)
    except Exception as e:
        print(f"Error loading the model: {str(e)}")
        print("Please ensure the model file exists and is in the correct format.")
        sys.exit(1)

    # Optimize stop-loss and take-profit
    results, best_result = optimize_sl_tp(loaded_model, X, data, confidence_threshold=0.3)

    # Run backtest with optimized parameters
    final_result = backtest(loaded_model, X, data,
                            confidence_threshold=0.3,
                            stop_loss=best_result['stop_loss'],
                            take_profit=best_result['take_profit'],
                            max_position_size=1.0,
                            min_position_size=0.1,
                            drawdown_limit=0.2)

    print("\nFinal Backtest Results:")
    print(f"Total Return: {final_result['total_return']:.2%}")
    print(f"CAGR: {final_result['cagr']:.2%}")
    print(f"Sharpe Ratio: {final_result['sharpe_ratio']:.2f}")
    print(f"Max Drawdown: {final_result['max_drawdown']:.2%}")
    print(f"Win Rate: {final_result['win_rate']:.2%}")
    print(f"Avg Trades per Year: {final_result['avg_trades_per_year']:.2f}")

    # Plotting the equity curve
    plt.figure(figsize=(12, 6))
    plt.plot(final_result['equity_curve'])
    plt.title(f'Equity Curve (SL: {final_result["stop_loss"]:.2f}, TP: {final_result["take_profit"]:.2f})')
    plt.xlabel('Trade Number')
    plt.ylabel('Equity')
    plt.show()

print("SPY daily data analysis completed.")
```
